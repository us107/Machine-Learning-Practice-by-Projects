```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

### Why

numpy → math & data

matplotlib → visualization

LinearRegression → same model for linear & polynomial

PolynomialFeatures → magic that creates x², x³, etc.

```python
np.random.seed(0) # random no. generator with reproducibility

X = np.linspace(-3, 3, 50).reshape(-1, 1) # 50 evenly spaced values between -3 and 3, reshaped to be a column vector
y = X**2 + 2 + np.random.randn(50, 1) # target variable with a quadratic relationship to X, plus some random noise
```

This cannot be learned by a straight line

```python
plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Non-linear Data")
plt.show()
```

**What this plot means (very short pointers):**

* X–Y points form a **U-shaped curve**
* Relationship is **non-linear**
* Lowest y values occur near **x = 0**
* y increases as |x| increases (both left and right)
* Noise is present → points don’t lie perfectly on a curve
* A **straight line cannot fit this pattern**
* Requires **polynomial (x²) term** to model curvature
* Classic example of **underfitting by linear regression**


```python
plt.scatter(X, y)

for degree in [1, 2, 4]:
    poly = PolynomialFeatures(degree=degree)
    X_poly = poly.fit_transform(X)
    
    model = LinearRegression()
    model.fit(X_poly, y)
    
    y_pred = model.predict(X_poly)
    plt.plot(X, y_pred, label=f"degree {degree}")

plt.legend()
plt.title("Effect of Polynomial Degree")
plt.show()
```

## Interpretation

Degree 1 → underfitting

Degree 2 → good fit

Degree 4 → risk of overfitting

Polynomial regression is NOT a new model.

It is:

Feature engineering (add x², x³…)

Linear regression on expanded features
